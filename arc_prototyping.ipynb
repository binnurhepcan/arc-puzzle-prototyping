```python
import os
import json
import matplotlib.pyplot as plt
import numpy as np
import inspect, numpy as np
```

```python
folder_path = "D:/py_projects/arc_kaggle/json_files"
json_files = []


for file in os.listdir(folder_path):
    if file.endswith(".json"):
        json_files.append(file)
        
print(json_files)
```

```python
index = 0
```

```python
file = json_files[index]

full_path = folder_path + "/" + file
print(full_path)

with open(full_path, "r") as f:   # "r" means read mode
    data = json.load(f) 

for task in data["train"][:5]:
    input_grid = task["input"] 
    output_grid = task["output"]
    plt.subplot(1, 2, 1)
    plt.imshow(input_grid)
    plt.subplot(1, 2, 2)
    plt.imshow(output_grid)
    plt.show()
```

```python
def solution(input_grid):
    arr = np.array(input_grid)
    mask = (arr > 0).astype(int)      
    op = np.kron(arr, mask)
    return op

for task in data["test"]:
    input_grid = task["input"] 
    output_grid = task["output"]
    test = solution(input_grid)
print(test)
print(output_grid)
print(np.array_equal(test, output_grid))
```

```python
if np.array_equal(test, output_grid):
    with open(f"solution{index:02d}.py", "w") as f:
        f.write("import numpy as np\n\n")
        f.write(inspect.getsource(solution))
    print("✅ Correct solution saved")
else:
    print("❌ Not saved")

index += 1
```
